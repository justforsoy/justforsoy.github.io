## 因果图和统计独立

因果关系侧重于一个节点的值是否会**导致另一个节点的值发生变化**，而统计关系侧重于了解节点的值是否**提供有关另一个节点的值的信息**。    
相关不等于因果，但是有因果图后，可以判断节点间是否条件独立。    

- 链式结构

```python
import graphviz as gr

g = gr.Digraph()

g.edge("X", "Y")
g.edge("Y", "Z")

g
```

如果Y值已观测，则X和Z独立（Conditional d-separation）     
$X \perp\!\!\!\perp Z \mid Y, P(Z | Y) = P(Z | Y, Z)$

- 分叉结构

```python
g1 = gr.Digraph()

g1.edge("X", "Y")
g1.edge("X", "Z")

g1
```

如果X值已观测，则Y和Z独立（Conditional d-separation）    
$Y \perp\!\!\!\perp Z | X, P(Z | X) = P(Z | X, Y)$

- 碰撞结构

```python
g2 = gr.Digraph()

g2.edge("X", "Z")
g2.edge("Y", "Z")

g2
```

如果Z值未观测，则X和Y独立（d-separation）。如果Z值已观测，则不独立    
$X \perp\!\!\!\perp Y$
