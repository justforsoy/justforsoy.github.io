“预测总是充满困难，尤其是对未来的预测”。这句话在荷兰政治家卡尔·克里斯蒂安·斯坦克在1940年代某个时候 allegedly 说出这句话时就是真的，并且在今天，特别是如果你正在处理时间序列和预测问题，这仍然是真的。    
时间序列分析有许多应用，从用预测做出预测，到理解历史趋势中的潜在因素。在本章中，我们将讨论一些贝叶斯方法来解决这个问题。我们将从将时间序列建模视为一个回归问题开始，设计矩阵是从时间戳信息解析的。然后，我们将探讨使用自回归组件模型时间相关性的方法。这些模型扩展到更广泛（更一般）的状态空间模型和贝叶斯结构时间序列模型（BSTS），我们将介绍线性高斯案例中的一种专门的推理方法：卡尔曼滤波器。本章的其余部分将简要总结模型比较以及选择时间序列模型的先验分布时需要考虑的因素。

## 6.1.时间序列问题总览

在许多实际应用中，我们按时间顺序观察数据，并在每次观察时生成时间戳。除了观察本身外，时间戳信息在以下情况下可能非常有用：    

- 存在时间**趋势**，例如地区人口、全球 GDP、美国每年的二氧化碳排放量。通常这是一个总体模式，我们可以直观地称之为“增长”或“下降”。

- 存在与时间相关的周期性模式，称为**季节性**。例如，每月温度的变化（夏季较高，冬季较低）、每月降雨量的变化（世界上许多地区冬季较低，夏季较高）、某个办公楼每天的咖啡消费量（工作日较高，周末较低）、每小时的自行车租赁数量（白天较高，夜晚较低），就像我们在第5章中看到的那样。

- 当前数据点以某种方式影响下一个数据点。换句话说，噪声或**残差**之间存在相关性。例如，每天在帮助台解决的案例数量、股票价格、每小时的温度、每小时的降雨量等。

因此将时间序列按以下分解为，是非常自然和有用的：
$$y_t = \text{Trend}_t + \text{Seasonality}_t + \text{Residuals}_t$$    

大多数经典的时间序列模型都基于这种分解。在本章中，我们将讨论对于展示一定程度时间趋势和季节性的时间序列的建模方法，并探索捕捉这些规律模式以及不太规则模式（例如，时间上相关的残差）的方法。
